1.1 Write a Python Program to implement your own myreduce() function which works exactly like
Python's built-in function reduce()

```python
from functools import reduce
```

```python
def myreduce(fun,l,init= None):
    iterable=iter(l)
    if init == None:
        result = next(iterable)
    else:
        result = init
    for i in iterable:
        result=fun(result,i)
    return result
```

```python
l=[1,2,3,4,5,6,7]
#print(l)
print(myreduce(lambda x , y :x+y,l))
print(myreduce(lambda x , y :x+y,l,10))
```

1.2 Write a Python program to implement your own myfilter() function which works exactly like
Python's built-in function filter()

```python
def myfil(fun,l):
    iterable=iter(l)
    for i in iterable:
        if fun(i):
            yield i
l=[1,2,3,4,5,6,7,8,9]
n=[]
for i in myfil(lambda x:x%2 !=0,l):
    #print (i)
    n.append(i)
print(n)
```

3. Implement a function longestWord() that takes a list of words and returns the longest one.


```python
def longestWord(l):
    length=[len(i) for i in l]
    initial=length[0]
    for i in range(1,len(length)):
        if length[i]>initial:
            initial=length[i]
    return l[length.index(initial)]
```

```python
l_1=["jhgfdas","zxvn,lk","qweykmnbjgh"]
longestWord(l_1)
```

1.1 Write a Python Program(with class concepts) to find the area of the triangle using the below
formula.
area = (s*(s-a)*(s-b)*(s-c)) ** 05
Function to take the length of the sides of triangle from user should be defined in the parent
class and function to calculate the area should be defined in subclass.

```python
class shape:
    def getSides(self):
        self.a = int(input('Enter length of side_1: '))
        self.b = int(input('Enter length of sid_2: '))
        self.c = int(input('Enter length of side_3: '))

class triangle(shape):
    def getArea(self):
        self.getSides()
        s = (self.a + self.b + self.c) / 2
        self.area = (s*(s-self.a)*(s-self.b)*(s-self.c)) ** 0.5
        print(f'area = {self.area}')
        
t1 = triangle()
t1.getArea()
```

1.2 Write a function filter_long_words() that takes a list of words and an integer n and returns the list
of words that are longer than n.

```python
lst=[]
n = int(input("Enter number of elements : "))  
for i in range(0, n): 
    ele = input() 
    lst.append(ele)
      
# print(lst)
def filter_long_words(l,n):
    return [i for i in l if len(i) > n]
        
l=int(input("enter the max length of word : "))
filter_long_words(lst,l)
```

2.2 Write a Python function which takes a character (i.e. a string of length 1) and returns True if it is
a vowel, False otherwise.


```python
def isvowel(c):
    if c in ['a', 'e', 'i', 'o', 'u','A', 'E', 'I', 'O', 'U']:
        return True
    else:
        return False
print(isvowel('j'))
print(isvowel('A'))
print(isvowel('y'))
```
